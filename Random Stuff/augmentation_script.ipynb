```python
from pydub import AudioSegment
import os

# Function to process audio files
def process_audio(input_folder, output_folder):
    # Iterate through genre folders
    for genre_folder in os.listdir(input_folder):
        genre_path = os.path.join(input_folder, genre_folder)

        # Initialize a counter for numbering segments within each genre
        segment_counter = 1

        # Create the output folder if it doesn't exist
        output_genre_folder = os.path.join(output_folder, genre_folder)
        os.makedirs(output_genre_folder, exist_ok=True)

        # Iterate through audio files in the genre folder
        for audio_file in os.listdir(genre_path):
            audio_path = os.path.join(genre_path, audio_file)

            # Load the audio file
            audio = AudioSegment.from_mp3(audio_path)

            # Calculate the middle section (30 seconds)
            middle_start = len(audio) // 2 - 15000  # 15000 milliseconds = 15 seconds
            middle_end = len(audio) // 2 + 15000   # 15000 milliseconds = 15 seconds
            middle_section = audio[middle_start:middle_end]

            # Ensure the audio is in mono format
            mono_section = middle_section.set_channels(1)

            # Divide the 30-second section into 10 segments of 3 seconds each
            segment_duration = 3000  # 3000 milliseconds = 3 seconds
            segments = [mono_section[i:i+segment_duration] for i in range(0, len(mono_section), segment_duration)]

            # Get the genre name from the parent folder
            genre_name = os.path.basename(genre_folder)

            # Save the segments as WAV files in the output genre folder with continuous numbering
            for i, segment in enumerate(segments):
                output_file = os.path.join(output_genre_folder, f"{segment_counter}.wav")
                segment.export(output_file, format="wav")
                print(f"Processing: {audio_file} - Segment {segment_counter}")
                segment_counter += 1

if __name__ == "__main__":
    input_folder = r"E:\Source_Code\PyWorld\ML Lab\Telegram"  # Replace with your input folder path
    output_folder = "E:\Research\ML - Machine Learning\Music Genre Classification\Final Data\wavs3sec"  # Replace with your output folder path

    process_audio(input_folder, output_folder)
    
    print("Audio file augmentation completed.")
```

```python
from pydub import AudioSegment
import os

# Function to convert MP3 files to WAV format and replace the original files
def convert_mp3_to_wav_in_place(input_folder):
    # Iterate through genre folders
    for genre_folder in os.listdir(input_folder):
        genre_path = os.path.join(input_folder, genre_folder)

        # Iterate through audio files in the genre folder
        for audio_file in os.listdir(genre_path):
            audio_path = os.path.join(genre_path, audio_file)

            # Check if the file is an MP3 file
            if audio_file.lower().endswith(".mp3"):
                # Load the MP3 file
                audio = AudioSegment.from_mp3(audio_path)

                # Generate the output file path with a .wav extension
                output_file = os.path.splitext(audio_path)[0] + ".wav"

                # Export the audio as a WAV file and replace the original file
                audio.export(output_file, format="wav")

                # Remove the original MP3 file
                os.remove(audio_path)

if __name__ == "__main__":
    input_folder = r"E:\Research\ML - Machine Learning\Music Genre Classification\Final Data\wavs30sec" # Replace with your input folder path

    convert_mp3_to_wav_in_place(input_folder)
    
    print("MP3 to WAV conversion completed. Original MP3 files are replaced.")
```

```python
from pydub import AudioSegment
import os

# Function to check if an audio file is corrupt
def is_corrupt_audio(file_path):
    try:
        AudioSegment.from_file(file_path)
        return False  # File is not corrupt
    except Exception as e:
        return True  # File is corrupt

# Root folder containing genre folders
root_folder = r"E:\Source_Code\PyWorld\ML Lab\Telegram"

# Iterate through genre folders
for genre_folder in os.listdir(root_folder):
    genre_path = os.path.join(root_folder, genre_folder)

    # Check if it's a directory
    if os.path.isdir(genre_path):
        print(f"Checking {genre_folder}...")

        # Iterate through audio files in the genre folder
        for audio_file in os.listdir(genre_path):
            audio_path = os.path.join(genre_path, audio_file)

            # Check if it's an MP3 file
            if audio_file.endswith(".mp3"):
                if is_corrupt_audio(audio_path):
                    print(f"Corrupt audio file: {audio_file}")

print("Corruption check completed.")
```
